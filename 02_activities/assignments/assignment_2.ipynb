```python
# To solve this problem, I need to first find the highest number in the given list. Then, I should generate a complete set of numbers ranging from 0 to that maximum value. Next, I will go through each number in this range and check if it is present in the original list. If a number is missing, I will add it to a separate list. Finally, if no numbers are missing, I will return -1; otherwise, I will return the list of missing numbers. Since this method involves looping through the full range and checking for missing values, it may not be the most efficient approach.
```


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
def find_missing_numbers(lst):
    max_value = max(lst)  
    
    full_range = set(range(0, max_value + 1))  
    missing_values = []
    
    for num in full_range:
        if num not in lst:
            missing_values.append(num)

    return missing_values if missing_values else [-1]  

# Example usage:
input_list = [0, 1, 2, 3, 4, 6, 7, 8]
find_missing_numbers(input_list) # Output: [5]


# Time Complexity:
# Finding the maximum value takes O(n)
# Creating the full range set takes O(n)
# Looping through and checking membership in a list takes O(n^2)
# Total time complexity would be O(n^2)

# Space Complexity:
# Creating the full range set takes O(n) space
# Storing the missing numbers takes O(n) space
# Total space complexity would be O(n)
```


-   Copy the solution your partner wrote. 


```python
from typing import List

def missing_num(nums: List[int]) -> List[int]:
    max_num = max(nums) 
    full = set(range(0, max_num + 1))  
    missing_numbers = list(full - set(nums))
    
    return missing_numbers if missing_numbers else -1  

missing_num([4, 6, 7, 8, 9]) # [0, 1, 2, 3, 5]
```


-   Explain why their solution works in your own words.


```python
# 1. The function first determines the highest number present in the input list. This ensures that we only check for missing numbers within a reasonable range.

# 2.  A set is created containing all numbers from 1 to the maximum value found in the list. This represents the complete sequence that should ideally exist if no numbers were missing.

# 3. The function converts the input list into a set and subtracts it from the full range set. This operation finds the numbers that exist in the complete sequence but are absent from the input list.

# 4. If there are missing numbers, they are returned as a list. Otherwise, the function returns -1 to indicate that no numbers are missing.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# The time complexity of this approach is O(n). First, finding the maximum number in the list requires scanning through all elements once, making it an O(n) operation. Next, constructing a set that includes all numbers from 1 to the maximum value also takes O(n) time. Lastly, determining the missing numbers by performing a set difference is another O(n) operation. Since each step runs in linear time, the overall time complexity remains O(n).

# For space complexity, the input list itself takes up O(n) space. Creating a full set of numbers from 1 to the maximum value also requires O(n) space. Additionally, storing the missing numbers in a separate set or list takes O(n) space in the worst case. Since all these components scale linearly with the input size, the total space complexity is O(n).
```
